```python
import pandas as pd
```

```python
data = pd.read_csv("2024 52 weak high.csv")
```

```python
# Step 2: Rename columns for consistency and clarity
data.columns = ['company_name', 'be_eq', '52_week_high_perf', '52_week_high_date', '52_week_low_perf', '52_week_low_date']
```

```python
# Step 3: Convert performance columns to numeric, handling errors for non-numeric values
data['52_week_high_perf'] = pd.to_numeric(data['52_week_high_perf'], errors='coerce')
data['52_week_low_perf'] = pd.to_numeric(data['52_week_low_perf'], errors='coerce')
```

```python
data['52_week_low_perf'].fillna(data['52_week_low_perf'].mean(), inplace=True)
```

```python
data['52_week_high_date'] = pd.to_datetime(data['52_week_high_date'], format='%d-%b-%y', errors='coerce')
data['52_week_low_date'] = pd.to_datetime(data['52_week_low_date'], format='%d-%b-%y', errors='coerce')
```

```python
print(data.head())
print(data.info())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
# Visualization 1: Bar plot of the 52-week high performance by company
plt.figure(figsize=(12, 6))
sns.barplot(x='company_name', y='52_week_high_perf', data=data.sort_values('52_week_high_perf', ascending=False).head(10))
plt.xticks(rotation=45)
plt.title('Top 10 Companies by 52-Week High Performance')
plt.xlabel('Company Name')
plt.ylabel('52-Week High Performance')
plt.show()
```
